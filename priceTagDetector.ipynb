```python
import cv2

model = YOLO("runs/detect/YOLO_8_2_S_150/weights/best.pt")

video_path = "raw_data/49_5/49_5.mp4"

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, conf=0.5, iou=0.9)

    annotated_frame = results[0].plot()

    scale = 0.5
    h, w = annotated_frame.shape[:2]
    resized = cv2.resize(annotated_frame, (int(w*scale), int(h*scale)))

    cv2.imshow("Annotated Frame", resized)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 

cv2.destroyAllWindows()
cap.release()
```

```python
import supervision as sv
from pathlib import Path

# Загружаем датасет
dataset = sv.DetectionDataset.from_yolo(
    images_directory_path="datasetv2/images",
    annotations_directory_path="datasetv2/labels",
    data_yaml_path="data.yaml"
)

# Разделяем на train/val
train_ds, val_ds = dataset.split(split_ratio=0.8, random_state=1)

# Создаем директории
Path("datasetv2/train/images").mkdir(parents=True, exist_ok=True)
Path("datasetv2/train/labels").mkdir(parents=True, exist_ok=True)
Path("datasetv2/val/images").mkdir(parents=True, exist_ok=True)
Path("datasetv2/val/labels").mkdir(parents=True, exist_ok=True)

# Экспортируем в YOLO формат
train_ds.as_yolo(
    images_directory_path="datasetv2/train/images",
    annotations_directory_path="datasetv2/train/labels"
)

val_ds.as_yolo(
    images_directory_path="datasetv2/val/images",
    annotations_directory_path="datasetv2/val/labels"
)
```

```python
# Progressive finetune: phase 1 (head) then phase 2 (full)
import torch
from ultralytics import YOLO

# Phase 1 - quick head finetune (faster, lower imgsz)
model = YOLO("runs/detect/YOLO_8_2_S_150/weights/best.pt")
results = model.train(
    data="data.yaml",
    epochs=20,
    imgsz=1024,
    batch=12,                     # уменьшите при OOM (например, 8 или 6)
    lr0=1e-4,                     # быстрый lr для адаптации головы
    device=0,
    degrees=5.0,
    translate=0.1,
    scale=0.3,
    shear=2.0,
    perspective=0.0005,
    fliplr=0.5,
    flipud=0.0,
    mosaic=0.5,
    mixup=0.05,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    close_mosaic=10,
)
```

```python

# Phase 2 - longer finetune at higher resolution
model = YOLO("runs/detect/YOLO_8_3_fine/weights/best.pt")  # или путь к чекпоинту из Phase 1
results = model.train(
    data="data.yaml",
    epochs=70,
    imgsz=1024,
    batch=4,                      # 8-12GB GPU → batch ~4–8; подгоните по OOM
    lr0=1e-5,                     # понижаем lr для стабильного дообучения
    device=0,
    degrees=5.0,
    translate=0.1,
    scale=0.3,
    shear=2.0,
    perspective=0.0005,
    fliplr=0.5,
    flipud=0.0,
    mosaic=0.5,
    mixup=0.05,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    close_mosaic=10,
)

# Быстрый инференс для проверки дубликатов (параметры NMS/Iou можно подправить)
preds = model.predict(source="raw_data/49_5/49_5.mp4", conf=0.5, iou=0.3, device=0)
print("Inference done; проверьте визуально и по метрикам.")
```
